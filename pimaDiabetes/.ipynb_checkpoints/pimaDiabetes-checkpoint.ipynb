```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
col_names = ['pregnant', 'glucose', 'bp', 'skin', 'insulin', 'bmi', 'pedigree', 'age', 'label']
pima = pd.read_csv("pimaDiabetes.csv", header=None, names=col_names)
pima.head()
```

```python
#Now, split the dataset into features and target variable as follows âˆ’
feature_cols = ['pregnant', 'insulin', 'bmi', 'age','glucose','bp','pedigree']
X = pima[feature_cols] # Features
y = pima.label # Target variable
```

```python
#70% training data and 30% of testing data 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
# Model with Classifier
clf = DecisionTreeClassifier(max_depth=3, random_state=0)
clf = clf.fit(X_train,y_train)
```

```python
# make new prediction
y_pred = clf.predict(X_test)
```

```python
# Accuracy, confusion matrix
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
result = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(result)
result1 = classification_report(y_test, y_pred)
print("Classification Report:",)
print (result1)
result2 = accuracy_score(y_test,y_pred)
print("Accuracy:",result2)
```

# Visualizing Decision Tree

```python
# plot with plot_tree
from sklearn import tree
from matplotlib import pyplot as plt
```

```python
# Visualise the tree
diabetes=plt.figure(figsize=(15,10))
_=tree.plot_tree(clf,filled=True,feature_names=X_train.columns)
# save to file
diabetes.savefig("diabetestree.png")
```
